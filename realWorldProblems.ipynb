```python
gas = pd.read_csv('/home/juto/Desktop/matplotlib/reference/matplotlib_tutorial/gas_prices.csv')
plt.figure(figsize=(10,5))

plt.title('USA Vs Canada Gas Prices', fontdict={'fontweight':'bold'})


plt.plot(gas.Year, gas.USA, "black", marker = 'o', label= 'USA')
plt.plot(gas.Year, gas.Canada, color='red', marker = '*', label= 'Canada')
plt.plot(gas.Year, gas['South Korea'], color='blue', marker = '^', label= 'South Korea')


plt.xticks(gas.Year[::3].to_list()+[2011])# adding the +[2011] adds to the list an extra year

plt.xlabel('Year')
plt.ylabel('USD')

plt.savefig('gas.png', dpi = 300)
plt.legend()
plt.show()

```
