```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

### Basic Graph

```python
# Line Number one
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize my graph
plt.figure(figsize=(9,4))

#plt.plot(x, y, label='2x', color= 'red', linewidth=2, marker='.', markersize=12)
#shorthand
plt.plot(x, y,"b--", label='2x', marker= '*')

# Line Number  Two
x2 = np.arange(0,4.5,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', marker= '*', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--', marker= ".")

plt.title('Our First Graph!', fontdict= {'fontsize' : 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```
