```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Graph

```python
# Line Number one
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize my graph
plt.figure(figsize=(9,4), dpi=100)

# Keyword arguments notation 
#plt.plot(x, y, label='2x', color= 'red', linewidth=2, marker='.', markersize=12)
#shorthand
plt.plot(x, y,"b--", label='2x', marker= '*')

# Line Number  Two
# select the interval we want to plot
"""
    The selected code snippet initializes an array 
    `x2` using the `np.arange()` function from the 
    NumPy library. This function generates a sequence
    of numbers starting from 0, ending at 4.5, and incrementing 
    by 0.5. The resulting array `x2` will contain the following 
    values: [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5]. 
    This array is used to plot part of the graph as a line in the 
    subsequent lines of code.
"""

x2 = np.arange(0,4.5,0.5)
"""
    The selected code snippet initializes an array `x2` using
    the `np.arange()` function from the NumPy library. This 
    function generates a sequence of numbers starting from 0, 
    ending at 4.5, and incrementing by 0.5. The resulting array
    `x2` will contain the following values: [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5].
    This array is used to plot part of the graph as a line in the subsequent lines of code.

In the context of the provided Python script, 
the `x2` array is used to plot a part of the graph as a 
line with a different style (red color, asterisk marker)
and then another part of the graph with a different style 
(red color, dashed line, dot marker). This technique is used 
to demonstrate how to plot different parts of a graph with 
different styles using the `plt.plot()` function in the 
Matplotlib library.
    """


# plots part of the graph as a line
plt.plot(x2[:6], x2[:6]**2, 'r', marker= '*', label='x^2')
# plots the remainder of the grapgh 
plt.plot(x2[5:], x2[5:]**2, 'r--', marker= ".")



# The graph
plt.title('Our First Graph!', fontdict= {'fontsize' : 20})
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])


plt.legend()
plt.savefig('mygrapgh.png', dpi=300)
plt.show()
```

Bar chart

```python
labels = ['A', 'B', 'c']
values = [1, 4, 2]

bars = plt.bar(labels, values[:len(labels)])


pattern = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))# if you dont put in the 0 it will run the patterns list from behind 
    bar.set_edgecolor('black')

bars[0].set_edgecolor('red')

plt.title('Bar Chart')
plt.xlabel('X Axis!')
plt.ylabel('Y Axis!')

    

"""
bars[0].set_hatch('/')
bars[1].set_hatch('\\')
bars[2].set_hatch('*')
"""

plt.savefig('barchart.png', dpi=300)
plt.figure(figsize=(6,4))
plt.show()
```
